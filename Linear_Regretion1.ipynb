```python
x = df[['Podiums']]
x.head()
```

```python
y = df['Points']
y.head()
```

```python
sns.scatterplot(x='Podiums', y='Points', data=df)
plt.title('Hubungan podium vs point')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
x_train.shape
```

```python

print(x_test.shape[0]/17, x_train.shape[0]/17)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.predict(x_test)
```

```python
model.predict(x_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
y_train_pred = model.predict(x_train)
y_train_pred
```

```python
error = y_pred - y_test
error
```

```python
error_train = y_train_pred - y_train
error_train
```

```python
print("Slope (Podiums):", model.coef_[0])
print("Intercept:", model.intercept_)
```

slope, ini menunjukan kemiringan garis. angka diatas menunjukan setiap kali seorang podium pointnya akan bertambah 25.7.
intercept, titik memotong garis subu Y, nilai prediksi point jka tidak pernah podium.

```python
from sklearn.metrics import mean_squared_error, r2_score
r2 = r2_score(y_test, y_pred)
r2
```

nilai r2 yang didapatkan 0.97. r2 mengukur proporsi variasi dalm variabel dependen yang dapat dijelaskan oleh variabel independen. yang berarti model sangat baik dalam menjelaskan variasi data.

```python
r2_train = r2_score(y_train, y_train_pred)
print("R-squared on training data:", r2_train)
print("R-squared on testing data:", r2)
```

Tidak ada indikasi underfitting: Model tidak underfit karena kinerja R-squared pada data pelatihan cukup tinggi.
Tidak ada indikasi kuat overfitting: Meskipun kinerja pada data pengujian sedikit lebih baik, perbedaannya tidak signifikan (terutama mengingat ukuran dataset yang kecil). Jika R-squared data pengujian jauh lebih rendah dari data pelatihan, baru itu akan menjadi tanda overfitting yang jelas.
